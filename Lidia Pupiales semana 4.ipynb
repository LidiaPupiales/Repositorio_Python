```python
#importar las librerías que se van a utilizar
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Elegimos la url de donde se está tomando los datos
url = "https://raw.githubusercontent.com/cienciadedatos/datos-de-miercoles/master/datos/2019/2019-04-10/partidos.txt"
```

```python
#Leemos el archivo con la librería pandas con el delimitador 
data_partidos = pd.read_csv(url, delimiter="\t")
```

```python
#Visualizamos la data cargada
print(data_partidos.head())
```

```python
partido_anio = data_partidos[(data_partidos.anio >= 1990)]
partido_anio
```

```python
partido_pais_argentina = partido_anio[(partido_anio.equipo_1 == 'Argentina')]
partido_pais_argentina
```

```python
partido_pais_uruguay = partido_anio[(partido_anio.equipo_1 == 'Uruguay')]
partido_pais_uruguay
```

```python
partido_pais_argentina.groupby('anio')
```

```python
partido_pais_uruguay.groupby('anio')
```

```python
partido_argentina_prom = partido_pais_argentina.groupby('anio')['equipo_1_final'].mean()
partido_argentina_prom
```

```python
partido_uruguay_prom = partido_pais_uruguay.groupby('anio')['equipo_1_final'].mean()
partido_uruguay_prom
```

```python
partido_argentina_prom = partido_argentina_prom.to_frame()
partido_argentina_prom
```

```python
partido_uruguay_prom = partido_uruguay_prom.to_frame()
partido_uruguay_prom
```

```python
partido_argentina_prom.index
```

```python
partido_uruguay_prom.index
```

```python
partido_argentina_prom.reset_index(inplace = True)
partido_argentina_prom
```

```python
partido_uruguay_prom.reset_index(inplace = True)
partido_uruguay_prom
```

```python
partido_argentina_prom.plot(x = 'anio', y = 'equipo_1_final')
plt.xlabel('Año')
plt.ylabel('Partido final Argentina')
plt.title('Resultados partidos de Argentina')
plt.savefig('Figura_Argentina.tiff')
partido_uruguay_prom.plot(x = 'anio', y = 'equipo_1_final')
plt.xlabel('Año')
plt.ylabel('Partido final Uruguay')
plt.title('Resultados partidos de Uruguay')
plt.savefig('Figura_Uruguay.tiff')
```

```python
fig = plt.figure()
eje = fig.add_subplot(1, 1, 1)
partido_argentina_prom.plot(x = 'anio', y = 'equipo_1_final', ax = eje, linestyle = '-.')
partido_uruguay_prom.plot(x = 'anio', y = 'equipo_1_final', ax = eje, linestyle = '--')
plt.xlabel('Año')
plt.ylabel('Promedio partidos')
plt.title('Resultados de partido -.-Argentina y ---Uruguay')
plt.savefig('Figura comparativo Argentina y Uruguay.tiff')
```
